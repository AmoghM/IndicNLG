```python
def extractData(path, title, values):
    res, titles = [], []
    with open(path, 'r', encoding="utf-8") as fr:
        for i in fr:
            
            row = eval(i)
            
            if row.get(title, None) is None:
                pass
            
            if row.get('what',None) == "मनुष्य":
                try:
                    temp = {value: row[value] for value in values}
                    res.append(temp)
                    titles.append(row[title])
                except KeyError:
                    pass

    fr.close()
    return res, titles
```

```python
def transformInfobox(infobox):
    box = ""
    for k,v in infobox.items():
        for pos, t in enumerate(indic_tokenize.trivial_tokenize(v)):
            box = box+k+"_"+str(pos+1)+":"+t+"\t"
    return box.rstrip("\t")

def getSummary(text, lang='hi'):
    sentences=sentence_tokenize.sentence_split(text, lang=lang)
    return sentences[0]
```

```python
def create_data_split(dataset):
    X, Y = [], []
    for data in dataset:
        if len(data['infobox'])>0 and len(data['extext'])>0:
            X.append(transformInfobox(data['infobox']))
            Y.append(getSummary(data['extext']))
    
    return X,Y
```

```python
X,Y = create_data_split(clean_merge)
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.1, random_state=42)
```

```python
def exportDataset(data, path):
    with open(path, "w+", encoding='utf-8') as fw:
        for item in data:
            fw.write(item+"\n")
    fw.close
```
