```python
wikidict = {}
unprocess_id = []
for i in range(1,len(root)):
    try:
        page_title = root[i][0].text
        page_text = root[i][3].find('{http://www.mediawiki.org/xml/export-0.10/}text').text
        wikidict[page_title] = page_text
    except AttributeError:
        unprocess_id.append(i)
        pass
```

```python
len(unprocess_id)
```

```python
wikidict_df = pd.DataFrame(list(wikidict.items()))
```

```python
wikidict_df.columns = ['Title', 'Text']
```

```python
wikidict_df.head()
```

```python
wikidict_df.to_csv('data/hindi_wikipedia.tsv', sep='\t', index= False)
```

```python
len(wikidict_df)
```

```python
infobox_df = wikidict_df[wikidict_df['Text'].str.contains("Infobox", case=False, na=False)]
```

```python
infobox_df.to_csv('data/hindi_wikibox.tsv',sep='\t',index=False)
```
